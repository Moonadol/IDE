```python
# Преобразуем столбец Date в формат datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

# Выделяем квартал из даты продажи
melb_df['Quarter'] = melb_df['Date'].dt.quarter

# Подсчитываем количество объектов, проданных в каждом квартале
quarter_counts = melb_df['Quarter'].value_counts()

# Находим второй по популярности квартал
second_most_popular_quarter = quarter_counts.sort_values(ascending=False).index[1]
second_most_popular_count = quarter_counts.sort_values(ascending=False).iloc[1]

print(f"Второй по популярности квартал: {second_most_popular_quarter}")
print(f"Количество объектов, проданных в этом квартале: {second_most_popular_count}")
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

# Максимальное количество уникальных значений для преобразования в тип category
max_unique_count = 150

# Преобразуем столбцы, в которых меньше 150 уникальных значений, в тип данных category
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')

# Подсчитываем количество столбцов, которые имеют тип данных category
category_columns_count = melb_df.select_dtypes(include=['category']).shape[1]

print(f"Количество столбцов, которые имеют тип данных category: {category_columns_count}")

# Выводим информацию о DataFrame
display(melb_df.info())

# Дополнительная проверка: выводим столбцы, которые были преобразованы в тип category
category_columns = melb_df.select_dtypes(include=['category']).columns
print("Столбцы, преобразованные в тип category:")
print(category_columns)
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
orders_products = orders_df.merge(
    product_df,
    left_on='ID товара',
    right_on='Product_ID',
    how='left'
)

# Проверка строк, где значения в столбце 'Name' отсутствуют
missing_product_info = orders_products[orders_products['Name'].isna()]

# Вывод строк с отсутствующими значениями в столбце 'Name'
print("Строки с отсутствующими значениями в столбце 'Name':")
print(missing_product_info)

# Вывод результирующей таблицы без строк с отсутствующими значениями в столбце 'Name'
final_orders_products = orders_products.dropna(subset=['Name'])
print("Результирующая таблица без строк с отсутствующими значениями в столбце 'Name':")
print(final_orders_products)
```
