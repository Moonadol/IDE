# Статистический анализ
Статистический анализ играет ключевую роль в современной медицине, помогая извлекать ценную информацию из огромного объема данных, накопленных в процессе исследований, диагностики и лечения. Статистика позволяет обрабатывать и интерпретировать эти данные, выявлять закономерности, оценивать эффективность различных методов лечения, определять факторы риска и делать прогнозы.
**Медицинская статистика** — это область статистики, которая специализируется на сборе, анализе и интерпретации данных, связанных с медицинскими исследованиями, заболеваниями, лечением и здоровьем человека. Она помогает врачам и исследователям принимать обоснованные решения на основе фактических данных и статистических закономерностей.
Медицинская статистика включает в себя различные аспекты:
* сбор данных,
* их классификацию,
* анализ,
* интерпретацию,
* визуализацию.

**Государственная медицинская статистика**
Государственная медицинская статистика представляет собой обширную область статистики, связанную со сбором, обработкой и анализом данных о здоровье населения, медицинской помощи и заболеваниях. В рамках государственной статистики выделяются несколько разделов:
* Эпидемиологическая статистика ― занимается анализом распространенности и распределения различных заболеваний среди населения, выявлением факторов риска и мер по их контролю. Эпидемиологическая статистика дает информацию о заболеваемости, смертности, вакцинации и других аспектах общественного здоровья.
* Статистика медицинских учреждений ― включает в себя информацию об активности больниц, поликлиник, амбулаторий, а также о качестве и объеме медицинской помощи, предоставляемой населению. Данные этого раздела помогают оценить доступность медицинских услуг, нагрузку на здравоохранение, эффективность работы медицинских учреждений.

* Статистика здоровья населения ― включает в себя информацию о физиологических показателях здоровья, образе жизни, привычках, смертности и продолжительности жизни. Эти данные позволяют проводить мониторинг здоровья населения, выявлять тенденции и проблемы в здравоохранении.

* Статистика медицинских работников ― этот раздел содержит информацию о квалификации, численности, распределении и оплате труда медицинского персонала. Данные о медицинских работниках необходимы для планирования кадрового состава в здравоохранении и обеспечения его эффективной работы.

* Статистика медицинского образования и научных исследований ― этот раздел включает данные о численности выпускников медицинских учебных заведений, направлениях исследований, научных публикациях и инновациях в медицине. Эти данные помогают понять состояние медицинской науки и образования, определить приоритеты развития и совершенствования медицинской отрасли.
**Виды медицинской статистики**
1. Дескриптивная (описательная) статистика используется для описания и суммирования данных без вывода общих заключений. Дескриптивная статистика позволяет представить информацию в виде числовых показателей, графиков или диаграмм, что облегчает восприятие и анализ данных в медицинских исследованиях.

2. Инференциальная статистика применяется для вывода общих заключений и обобщений на основе анализа выборочных данных. Инференциальная статистика играет важную роль в проведении клинических исследований и определении эффективности лечения, поскольку позволяет делать обобщения о целой популяции на основе данных, собранных из выборки.

3. Биостатистика специализируется на анализе данных, связанных с биологическими и медицинскими исследованиями. Биостатистика включает в себя методы анализа клинических испытаний, эпидемиологических исследований, генетических данных и других областей медицины, где требуется статистическая обработка информации.

4. Эпидемиологическая статистика направлена на изучение распространения болезней в популяции, выявление факторов риска и оценку воздействия общественного здравоохранения. Эпидемиологическая статистика используется для анализа эпидемиологических данных, создания моделей распространения заболеваний и разработки мероприятий по их контролю и профилактике.

# Генеральная совокупность, выборка
Объектом исследования в статистике является статистическая совокупность ― группа, состоящая из большого числа относительно похожих единиц (больных каким-либо заболеванием, случаев неудовлетворенности медицинской помощью, пациентов, получивших инвалидность, случаев впервые возникшего заболевания).
**Генеральная совокупность** ― все изучаемые объекты того вида, над которым исследователи проводят наблюдения в целях получить конкретные значения случайной величины. Например, если мы тестируем новое лекарство от некоторого заболевания, то генеральной совокупностью для нас будут абсолютно все пациенты, страдающие данным заболеванием.
Часто на практике ученые работают только с небольшим количеством случайно отобранных объектов из изучаемой генеральной совокупности, называемых выборкой (выборочной совокупностью).

*Пример*
Цель исследования ― изучить здоровье студентов Сеченовского университета.

Генеральная совокупность ― все студенты Сеченовского университета на данный момент. В 2024 году объем совокупности составляет примерно восемнадцать тысяч человек (N=18 000).

Выборочная совокупность ― двести студентов, обучающихся в Сеченовском университете в 2024 году. (n=200).

**Репрезентативность**
Обязательным условием выборочной совокупности является репрезентативность (от англ. representative ― представительный), когда все объекты выборки хорошо отражают изучаемые свойства генеральной совокупности.

Выборка должна быть репрезентативной в плане количества и качества.
**Количественная репрезентативность** связана с объемом выборки и обеспечивает статистическую значимость результатов исследования. Следовательно, чем больше объем выборки, тем более достоверными будут полученные результаты.

Пример количественной репрезентативности можно увидеть в клинических исследованиях, где большое количество пациентов включается в исследование. Это позволяет сделать выводы, применимые к более широкой популяции.
**Качественная репрезентативность** связана с тем, насколько хорошо выборка представляет целевую популяцию. Это значит, что выборка должна быть представительной и разнообразной, чтобы ее характеристики отражали популяцию. Это позволит обобщить результаты исследования на всю целевую группу.

Примером качественной репрезентативности может служить исследование о побочных эффектах лекарства у пожилых людей. Если выборка состоит из представителей разных возрастных групп, включая тех, кто страдает хроническими заболеваниями, проживающих как в городе, так и в сельской местности, то результаты такого исследования будут более качественно репрезентативными и позволят делать более обоснованные выводы о безопасности лекарства для всей группы пожилых пациентов.
Существует несколько **методов для формирования репрезентативной выборки**, представляющих различные подходы к отбору участников исследования:
* Случайная выборка ― каждый элемент популяции имеет одинаковые шансы попасть в выборку. С помощью случайной выборки можно обеспечить объективность результатов и минимизировать вероятность искажения данных.
* Стратифицированная выборка ― популяция делится на различные страты (группы) по определенным признакам, а затем из каждой страты отбирается участники выборки. Этот метод позволяет учесть гетерогенность популяции и обеспечить равное представительство различных групп.
* Кластерная выборка ― популяция делится на кластеры (например, районы, школы, предприятия), из каждого кластера случайным образом выбирается определенное количество участников исследования. Этот метод удобен при работе с большими популяциями и позволяет сократить затраты на сбор информации.
* Удобственная выборка ― участники выбираются на основе удобства доступа или желания принять участие в исследовании. Этот метод не всегда обеспечивает репрезентативность выборки, так как участники могут существенно отличаться от общей популяции.
* Квотная выборка ― участники выборки отбираются на основе заранее заданных квот по различным признакам (например, полу, возрасту, образованию). Этот метод позволяет обеспечить баланс по различным характеристикам, однако требует внимательного контроля за процессом отбора участников.
**Эффективное проведение статистического исследования**
Статистические исследования в медицине играют ключевую роль в обеспечении эффективности и надежности медицинских практик. Они позволяют оценить эффективность лекарственных препаратов, выявить факторы риска развития заболеваний, определить влияние генетических и окружающих факторов на здоровье человека, а также принять обоснованные решения в области здравоохранения.

Эффективное проведение статистического исследования в медицине включает в себя несколько ключевых этапов.

1. Определение проблемы исследования

Проблемой в общественном здоровье и здравоохранении могут быть:

недоступность качественной медицинской помощи из-за отсутствия медицинских учреждений, длительных очередей на прием к врачу или недостаточного количества специалистов;
эпидемии и пандемии, требующие масштабных медицинских и предупредительных мер;
проблемы психического здоровья населения;
недостаточное финансирование здравоохранения, приводящее к нехватке медицинского оборудования, медикаментов и низким зарплатам медицинского персонала;
антимикробная резистентность, затрудняющая лечение инфекционных заболеваний и требующая разработки новых подходов к борьбе с этим явлением;
проблемы вакцинации в некоторых регионах, увеличивающие риск возникновения вспышек инфекционных заболеваний, которые могли бы быть предотвращены с помощью прививок.
2. Формулирование задач исследования

На этом этапе определяются цель и задачи исследования, формулируются гипотезы, которые необходимо подтвердить или опровергнуть. Четкость постановки задач обеспечивает правильное планирование и проведение исследования.

3. Определение методологии исследования

Выбираются методы сбора данных, разрабатывается план исследования, определяются критерии включения и исключения пациентов, а также проводится расчет объема выборки.

4. Сбор данных

Проведение статистического исследования включает в себя сбор разнообразной информации о пациентах или группах, которые подвергаются анализу. Этот этап требует тщательного и строго контролируемого сбора данных.

5. Обработка данных

После сбора данных происходит их структурирование, кодирование и ввод в статистические программы для анализа. На этом этапе используются статистические методы для выявления в данных закономерностей и тенденций.

6. Анализ результатов

После обработки данных проводится анализ полученных результатов с использованием статистических методов. На основе анализа делаются выводы и формулируются рекомендации для практического применения.

7. Интерпретация результатов и формулирование выводов

Полученные данные помогают принимать важные решения в области медицины, разрабатывать новые методики лечения, профилактики и диагностики.
*Пример*
**Проблема:** в некоторых районах города наблюдается высокий уровень заболеваемости астмой среди подростков.

**Задача:** провести статистическое исследование с целью выявить возможные факторы, влияющие на высокий уровень заболеваемости астмой среди подростков в указанных районах.

**Цель:** проанализировать связь между экологическими условиями, образом жизни, семейным положением и развитием астмы среди подростков.

**Гипотеза:** высокие показатели загрязнения окружающей среды, неблагоприятные условия проживания, отсутствие спортивной активности и наличие неблагополучной семейной ситуации способствуют увеличению риска развития астмы среди подростков в определенных районах города.

**Определение методологии исследования:** выбор методов сбора данных, разработка плана исследования, определение критериев включения и исключения подростков.

**Сбор данных:** сбор информации о заболеваемости астмой, экологических условиях, образе жизни и семейных факторах у подростков.

**Обработка данных:** структурирование, кодирование и анализ данных с использованием статистических программ.

**Анализ результатов:** анализ полученных данных, проверка гипотезы, выявление закономерностей и выводов о влиянии различных факторов на заболеваемость астмой среди подростков в районах города.

**Подготовка отчета:** формулирование выводов и рекомендаций на основе полученных результатов и подготовка научного отчета для дальнейшего запроса.
# Признаки
* Единица наблюдения ― конкретный объект или субъект, на который направлено внимание и изучение в рамках исследования. Это может быть пациент, студент, врач, который становится объектом анализа и изучения для получения информации о его характеристиках, свойствах и поведении.

* Признаки сходства ― характеристики или аспекты, общие у различных объектов или единиц наблюдения. Эти признаки позволяют выявить сходство между объектами и установить общие черты или свойства, которые характеризуют их как однородные или схожие, с точки зрения определенных параметров.

* Признаки различия ― свойства или характеристики, по которым объекты или единицы наблюдения различаются. Эти признаки помогают выделить индивидуальные особенности и уникальные аспекты каждого объекта, позволяя провести сравнение и выявить различия между ними.

* Учетные признаки ― параметры, характеристики или переменные, которые используются для учета и описания объектов или единиц наблюдения. Эти признаки могут быть как качественными, так и количественными. Служат для систематизации и классификации данных, а также для последующего анализа и интерпретации результатов исследования.
**Классификация признаков по характеру: качественные, количественные**
**Количественные признаки** представляют собой числовые значения, которые могут быть измерены и подвергнуты математическим операциям. О количественных мы можем сказать: «на три единицы меньше», «в шесть раз больше». Они позволяют более точно описывать и анализировать данные, поскольку представляют собой конкретные численные значения. Количественные признаки могут быть непрерывными или дискретными.

**Непрерывные признаки** могут принимать любые значения в определенном диапазоне. Например, температура, рост пациента, биохимические показатели крови. Над непрерывными признаками можно проводить арифметические операции.

**Дискретные признаки**, напротив, принимают конкретные значения, такие как количество детей в семье, число рецидивов, число заболеваний у одного больного, число вызовов скорой помощи, число пациентов, поступающих в приемное отделение городской больницы.

Использование количественных признаков позволяет проводить более глубокий анализ данных, выявлять закономерности, строить математические модели и делать прогнозы.

Качественные признаки ― свойства объектов, которые характеризуются не числовыми значениями, а категориями, метками или описательными обозначениями. Они описывают качественные или категориальные характеристики объекта и определены через их наличие или отсутствие, принадлежность к определенной группе или категории.

**Качественные признаки** позволяют классифицировать объекты по определенным свойствам и качествам, что позволяет исследователям делать выводы о сходствах и различиях между объектами в категориальном аспекте.

Например, способ лечения: хирургический, радиотерапия, обсервационный. Такие признаки ещё называют категориальными, поскольку их можно отнести к той или иной категории.

Среди категориальных признаков можно выделить:

**Номинальные признаки** ― это переменные, значения которых не имеют внутреннего порядка, а просто обозначают различные категории или качественные характеристики. В медицине номинальные признаки играют важную роль при классификации пациентов, диагнозе заболеваний и определении других особенностей.

Примеры номинальных признаков в медицине можно найти в различных областях, таких как эпидемиология, диагностика, лечение и мониторинг пациентов:

* Тип заболевания ― инфекционное, онкологическое, аутоиммунное или другое. Номинальный признак «тип заболевания» помогает врачам классифицировать и выбирать соответствующие методы лечения.
* Группа крови человека ― номинальный признак, который может быть полезен для применения при трансфузии крови, определении совместимости доноров и реципиентов, а также в генетических исследованиях.
* Тип вакцинации ― при проведении вакцинации определяется тип вакцины, которая была использована (вирусная, бактериальная, деленная, живая или мертвая). Этот номинальный признак важен для отслеживания эффективности вакцинации и реакции пациента на прививку.
* Пол пациента ― часто используется как номинальный признак для сегментации пациентов по половому признаку. Это может быть важно при анализе различий в заболеваниях, реакции на лечение и других параметрах.
* Методы диагностики ― рентгенологические, лабораторные, исследования с использованием ультразвука и томографии. Это помогает идентифицировать, какие методы были использованы для диагностики конкретного заболевания.
**Дихотомические (бинарные)** признаки представляют собой категориальные переменные, которые могут принимать только два возможных значения. Такие признаки делят объекты на две группы в зависимости от наличия или отсутствия определенного свойства.
**Порядковые (ранговые, ординальные) признаки** представляют собой переменные, значения которых имеют внутренний порядок, но не обязательно равные интервалы между ними. Такие признаки позволяют упорядочить объекты по степени или уровню наблюдаемого свойства, но не позволяют проводить математические операции с точными числовыми значениями. Порядковые признаки широко используются в медицине для оценки тяжести заболеваний, уровня боли, степени нарушения функций органов и других параметров:

* Шкала боли (нумерическая шкала оценки боли от 1 до 10). Врачи могут попросить пациента оценить свою боль по шкале от 1 (без боли) до 10 (наивысшая возможная боль). Такая шкала позволяет оценить интенсивность болевого синдрома и принять решение о необходимом лечении.
* Стадии заболевания (например, стадии рака). Часто заболевания классифицируются на различные стадии в зависимости от распространенности процесса. Например, рак может иметь стадии от I до IV, где более высокая стадия указывает на более продвинутый и тяжелый случай заболевания. Такая классификация помогает врачам определить тактику лечения и прогноз для пациента.
* Уровни образования (например, начальное, среднее, высшее). Данный порядковый признак может использоваться при исследованиях в области общественного здравоохранения для анализа влияния образования на здоровье пациентов. Уровень образования может быть связан с доступом к медицинской информации, привычками ведения здорового образа жизни и другими факторами, влияющими на состояние здоровья.
**Классификация признаков по роли в совокупности**: факторные, результативные
По роли в совокупности выделяют факторные признаки ― переменные, которые могут влиять на результативный признак. То есть на конечный результат или исход, который мы хотим изучить или прогнозировать. Таким образом, факторные признаки имеют причинно-следственную связь с результативными показателями.
# Абсолютные и относительные величины. Интенсивный показатель
Описательная статистика в медицине играет ключевую роль в анализе и интерпретации данных. Этот метод позволяет систематизировать, описывать и представлять информацию о популяции пациентов, их заболеваниях, лечении и других медицинских параметрах.

Описательная статистика имеет несколько ключевых задач:

1. Описание и классификация данных

Включает в себя вычисление мер центральной тенденции (среднее, медиана, мода) и мер изменчивости (стандартное отклонение, диапазон, квартили).

2. Визуализация данных

Создание графиков, диаграмм и других визуальных представлений данных. Это помогает выявить закономерности, аномалии и взаимосвязи между переменными.

3. Сводка данных и выявление основных характеристик

Сведение данных в краткую форму, выделение основных характеристик выборки или популяции помогает исследователям быстрее понимать сущность данных и делать выводы.

4. Сравнение групп и распределений

Описательная статистика позволяет сравнивать различные группы пациентов или факторы в выборке для выявления статистически значимых различий. Это особенно важно при проведении медицинских исследований и анализе эффективности лечения.

5. Выявление ассоциаций и корреляций

Выявление связи и корреляции между переменными позволяет понять влияние одних факторов на другие. Это полезно при исследовании медицинских данных и их взаимосвязей.

В процессе анализа медицинских данных используются абсолютные и относительные величины.

**Абсолютные величины** используются для описания фактических числовых значений данных, например, количество пациентов с определенным диагнозом или число лекарственных препаратов, выписанных в определенный период.

**Относительные величины** позволяют сравнивать данные в различных контекстах и на разных уровнях. Например, относительное количество пациентов с определенным заболеванием может быть выражено в процентном соотношении к общему количеству пациентов в клинике. Это помогает оценить распространенность заболевания и эффективность лечения.

Различают четыре основных вида относительных величин:

* Экстенсивные величины, такие как объем крови, масса тела, количество кровяных клеток, зависят от объема организма.
Например, общая масса крови в организме человека является экстенсивной величиной, так как ее значение изменяется в зависимости от общего объема кровообращения в организме.

* Соотношения позволяют выявлять взаимосвязи между различными параметрами организма.
Например, отношение количества лейкоцитов к общему числу клеток крови может указать на наличие воспалительного или инфекционного процесса в организме. Такие соотношения помогают диагностировать различные заболевания и контролировать их лечение.

* Наглядные величины в медицине, такие как индекс массы тела, коэффициенты роста и веса, служат для визуализации и оценки физического состояния человека.
Например, использование индекса массы тела позволяет оценить соответствие между массой тела и ростом человека, что важно для определения возможного риска ожирения или дефицита веса.

* Интенсивный показатель ― ключевой инструмент для диагностики различных состояний здоровья человека. Этот тип показателя обозначает характеристику, которая не зависит от общего объема организма, а отражает уровень определенного физиологического процесса. В медицине интенсивные показатели могут быть использованы для оценки работы органов и систем, контроля за состоянием пациента, а также для мониторинга эффективности лечения.
Например, пульс, давление, температура тела ― это типичные интенсивные показатели, которые активно используются в клинической практике. Измерение пульса позволяет оценить работу сердечно-сосудистой системы, выявить аритмии, недостаточность кровообращения и другие заболевания. Анализ показателей давления помогает контролировать работу сердца, обнаруживать гипертонию, гипотонию. Температура тела также является важным интенсивным показателем, позволяющим определить наличие воспалительных процессов, инфекций или других патологий.
# Экстенсивный показатель
**Экстенсивный показатель** — относительная величина, показывающая, как велика отдельная часть по отношению ко всей изучаемой совокупности. Одним из основных экстенсивных показателей, используемых в медицинских исследованиях, является общее количество заболевших в определенной популяции или общее количество случаев заболевания в определенном временном периоде.
Классификация экстенсивных показателей:

* Показатели удельного веса части в целом. Например, доля заболевших короновирусом среди всех заболеваний.
* Показатели структуры или распределения. Распределение всех заболеваний, зафиксированных за год, по отдельным заболеваниям.

**Расчет экстенсивных показателей:**

1. Процент заболеваемости в определенной популяции

Для определения процента заболеваемости в популяции можно рассчитать отношение числа заболевших к общему числу людей в этой популяции и умножить на 100%.

Например, если в популяции из 1000 человек было 50 заболевших, то процент заболеваемости составит (50 / 1000) * 100% = 5%.

2. Общее количество случаев заболеваний

Для определения экстенсивного показателя общего количества случаев заболеваний в популяции за определенный период времени нужно просто сложить все единичные значения количества заболевших за каждый месяц или год.

Например, если у нас есть данные о числе заболевших в популяции каждый месяц за год, то, суммируя все значения, мы получим общее количество случаев заболеваний за весь год.

3. Суммарное количество проведенных операций

Для оценки экстенсивного показателя суммарного количества проведенных операций в больнице за год можно сложить все значения операций за каждый месяц.

Например, если в январе было проведено 100 операций, в феврале 120 операций и так далее, то общее количество операций за год будет равно сумме этих значений.
# Показатель соотношения
Характеризует соотношение между двумя совокупностями, не связанными между собой (соотношение абортов и родов, обеспеченность населения койками, врачами, соотношение врачей и другого медицинского персонала).

Чтобы вычислить показатель соотношения, нужны две совокупности.

# Показатель наглядности
Используется в анализе однородных чисел и применяется, когда необходимо «уйти» от показа истинных величин (относительных и средних величин, абсолютных чисел). Например, когда проводится сравнительный анализ одних и тех же показателей, но на разных территориях или в разное время. Часто эти величины представлены в динамике.

Чтобы вычислить показатель наглядности, одну из сравниваемых величин принимают за 100% (часто исходную величину). Остальные вычисляются в процентном отношении к ней.
# Показатели описательной статистики
Показатели описательной статистики можно разбить на три основные группы:

* Меры центральной тенденции ― статистические показатели, которые используются для описания среднего значения или «средней точки» распределения данных. Они представляют собой характеристики, которые позволяют суммировать или усреднять данные в выборке для представления их общего положения или среднего значения.
Меры центральной тенденции помогают сжать разнообразные данные в компактный и информативный числовой показатель, который отражает основные характеристики распределения и позволяет делать выводы о типичном или среднем значении в исследуемой выборке.

Меры центральной тенденции: среднее, мода, медиана.

* Меры разброса (изменчивости) ― характеризуют различия между отдельными значениями выборки. Используются для измерения того, насколько данные распределены вокруг их среднего значения и позволяют судить о степени однородности полученного множества, его компактности.
Меры изменчивости: размах, дисперсия, стандартное отклонение.

* Графические представления результатов ― гистограмма, частотная диаграмма и другие.
**Медиана** ― статистическая мера центральной тенденции, которая представляет собой серединное значение в упорядоченном ряду данных. Другими словами, медиана разделяет упорядоченный набор данных на две равные части: одна половина значений расположена выше медианы, а другая ― ниже.
Для расчета медианы необходимо:

1.  Упорядочить данные по возрастанию или убыванию.

2. Если количество данных нечетное, то медиана будет равна значению данных, находящемуся после упорядочивания посередине.
Где n ― общее число значений признаков в ряду.

3. Если количество данных четное, то медиану вычисляют как среднее арифметическое двух центральных значений.


Таким образом, медиана является важной статистической мерой, которая помогает определить центральное значение в наборе данных и устойчива к выбросам, что делает ее полезной для анализа даже при наличии аномальных значений.
**Мода** ― значение или значения в наборе данных, которые встречаются наиболее часто. Другими словами, мода представляет собой наиболее часто встречающееся значение или значения в выборке. Мода является одним из показателей центральной тенденции и может быть полезна при анализе распределения данных.
Чтобы рассчитать моду, нужно:

1. Проанализировать набор данных и определить, какие значения встречаются наиболее часто.

2. Выбрать значение или значения, которые встречаются чаще всего. Если в наборе данных есть несколько значений, которые встречаются одинаковое количество раз и чаще всех остальных, то такие значения называются модальными значениями.

Если в наборе данных нет однозначно выраженного модального значения (если все значения встречаются одинаковое количество раз), то такой набор данных называется унимодальным.

Мода может быть полезна при анализе данных, особенно в случаях, когда необходимо выявить наиболее распространенные или популярные значения. Однако важно помнить, что мода может не выражать полную картину распределения данных, так как не учитывает все значения в выборке.
**Симметричное распределение** ― когда большинство наблюдений в центре распределения, и чем дальше мы отклоняемся от середины распределения, тем реже встречаются значения признака.
# Параметры распределения
**Параметры распределения** ― это числовые характеристики распределения вероятностей случайной величины, которые определяют его форму, положение и масштаб. Параметры позволяют описывать распределение данных и делать выводы о его свойствах.
Каждое распределение имеет свои уникальные параметры, которые определяют его форму и характеристики. Наиболее важными параметрами являются:

* среднее значение (среднее арифметическое),
* дисперсия,
* стандартное (среднеквадратическое) отклонение,
* математическое ожидание,
* показатели эксцесса,
* показатели асимметрии.
**Среднее**
Если мы расположим женщин и мужчин на одной шкале роста, то увидим, что распределение женщин ниже, чем распределение мужчин.

Характеристика положения распределения на числовой оси называется средним.

Среднее по генеральной совокупности обозначается греческой буквой μ (читается «мю») и вычисляется как отношение суммы значений признака для всех членов совокупности к числу членов совокупности.

**Среднее арифметическое** ― это то значение величины, которое получается, когда сумма всех наблюдаемых значений распределяется поровну между всеми единицами наблюдения.

Например, вычислив среднее арифметическое количества принятых в поликлинике пациентов за месяц, мы найдем то количество больных в день, которые обратились бы за помощью, если бы каждый день заболевало одинаковое число людей.

Рассмотрим базовые свойства выборочного среднего:
 
**Стандартное отклонение, дисперсия**
Стандартное (среднеквадратическое) отклонение измеряет разброс значений в выборке относительно их среднего значения (в тех же единицах измерения, что и сама переменная) и позволяет оценить степень изменчивости данных. Чем больше значение стандартного отклонения, тем больший разброс данных можно наблюдать.
**Дисперсия**
Это мера разброса или вариации значений в наборе данных относительно их среднего значения. Вычисляется нахождением среднего квадратического отклонения от среднего значения.
Для вычисления дисперсии необходимо выполнить несколько шагов:

1. Сначала находим разность между каждым значением генеральной совокупности (выборки) и средним значением всей генеральной совокупности (выборки).
2. Затем эти разности возводим в квадрат, чтобы учесть только положительные значения и избежать компенсации отклонений относительно среднего.
3. После этого сумму квадратов разностей делим на количество значений в генеральной совокупности (количество значений в выборке минус одно).
