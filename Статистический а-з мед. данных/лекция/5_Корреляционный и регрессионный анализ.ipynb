# **Введение**
**Корреляционный и регрессионный анализ**
Корреляционный анализ позволяет оценить силу и направление взаимосвязи между двумя или более переменными, помогая определить, есть ли статистически значимая зависимость между ними: насколько сильно одна переменная связана с другой, что может быть полезно для выявления факторов, влияющих на здоровье пациентов или эффективность лечения.

В медицине корреляционный анализ широко используется для исследования различных биомедицинских данных, оценки эффективности лечения, оптимизации диагностики и мониторинга пациентов, для исследования взаимосвязей между клиническими показателями, результатами лабораторных анализов, показателями здоровья пациентов и другими важными факторами.

Регрессионный анализ позволяет строить прогностические модели, предсказывающие значения зависимой переменной, на основе одной или нескольких независимых переменных.

Этот метод анализа данных широко используется в медицинских исследованиях для прогнозирования и оценки эффективности лечения, выявления факторов риска заболеваний, а также оптимизации медицинских процессов.
# **Функциональные зависимости**
Функциональная зависимость между переменными  и  означает, что каждому значению одной переменной соответствует строго определенное значение другой переменной.
Исследования в области медицины часто сталкиваются с ситуациями, когда нет чёткой или строгой функциональной зависимости между различными медицинскими переменными. Этот феномен может быть обусловлен многочисленными факторами, вот основные:

* Уникальные особенности каждого человека. Генетические различия, степень физической активности, питание, окружающая среда — все эти факторы могут влиять на результаты исследований и усложнять установление однозначных функциональных зависимостей.
* Статистические особенности данных. Иногда нехватка данных, неточности измерений или различия в методиках исследования могут затруднить выявление чётких функциональных зависимостей между медицинскими переменными.
* Биологическая система человека представляет собой сложный и многоуровневый механизм, в котором взаимодействуют множество переменных. В нейрофизиологии, например, даже при изучении взаимосвязей между физиологическими показателями и психологическими процессами не всегда удается установить прямую функциональную связь.
**Статистическая зависимость** — это взаимосвязь между двумя или более переменными, которая проявляется в изменениях одной переменной при изменениях другой или других переменных.
**Корреляционная зависимость** — это статистическая взаимосвязь между двумя или более переменными, при которой изменения в одной переменной сопровождаются определенными изменениями в другой переменной.
**Корреляция** — это мера степени наблюдаемой связи между переменными, но сам по себе этот статистический показатель не раскрывает причинности.
Основные задачи корреляционного анализа:
1. Оценка степени и направления связи между переменными

Корреляционный анализ позволяет определить, насколько сильно связаны между собой две или более переменные, и выявить направление этой связи (положительное или отрицательное).

2. Предсказание значений переменных

На основе корреляционного анализа можно строить модели прогнозирования, которые позволяют предсказывать значения одной переменной на основе значений другой переменной.

3. Выявление скрытых зависимостей

Корреляционный анализ может помочь выявить скрытые взаимосвязи между переменными, которые могут быть полезными для понимания механизмов взаимодействия между изучаемыми явлениями.

4. Идентификация влиятельных переменных

Анализ корреляций позволяет определить, какие переменные оказывают наибольшее влияние на изменения других переменных, что может быть полезно при принятии решений.
# **Связь**
Исследование взаимосвязей между различными переменными в медицине представляет собой важное направление, которое позволяет выявлять и понимать связи между различными заболеваниями, факторами риска, лечением и исходами. Врачи и исследователи сталкиваются с необходимостью различать и анализировать существующие связи между показателями здоровья и влияющими на них факторами.
*Примеры неверных выводов, основанных на наблюдениях:*
* Существует ярко выраженная взаимосвязь между результатами теста по грамматике русского языка и ростом человека. Это можно объяснить тем, что у лиц с более высоким ростом больше объем мозга, что способствует лучшему запоминанию правил грамматики и лингвистических норм.
* Люди, употребляющие большое количество витамина C, реже заболевают простудой.
* Уровень рождаемости в регионе тесно коррелирует с числом аистов, обитающих на данной территории. Из этого можно сделать вывод, что именно аисты приносят детей.
* Люди, употребляющие большое количество кофеина, имеют повышенное кровяное давление.
*Верные объяснения:*
* Тест провели у школьников без учета их возраста. С возрастом у школьников увеличивается не только рост, но и знание русского языка. Таким образом, наблюдаемая взаимосвязь между результатами теста и ростом обусловлена третьим фактором — возрастом.
* Заключение о том, что витамин C укрепляет здоровье, может оказаться ошибочным потому, что эти люди могут в целом вести более здоровый образ жизни или уделять больше внимания своему здоровью.
* Уровень рождаемости исторически всегда оказывался выше в сельской местности, а не в городах. При этом аисты чаще проживают именно на сельских территориях. Таким образом, связь между данными явлениями может быть объяснена воздействием второго фактора — типом местности.
* Есть другие факторы, например, стресс или наследственность, которые влияют на кровяное давление сильнее, чем кофеин.
# **Диаграмма рассеяния**
Диаграмма рассеяния — это важный инструмент визуализации данных, который позволяет исследовать отношения между двумя переменными. Этот графический метод позволяет наглядно представить распределение точек данных по плоскости и выявить какие-либо закономерности или зависимости.
Для более полного представления о взаимосвязях между переменными целесообразно создавать матрицу диаграмм рассеяния, которая позволяет визуализировать взаимосвязи между несколькими переменными одновременно, что помогает исследователям выявлять скрытые закономерности, тенденции и аномалии в данных.
# **Коэффициент корреляции**
Для проведения измерения и оценки статистической значимости корреляционной связи применяют коэффициент корреляции () и его среднюю ошибку ().

Коэффициент корреляции — это числовая характеристика, которая позволяет одним числом характеризовать силу взаимосвязи между исследуемыми явлениями и направление взаимосвязи между ними.
Существуют различные типы корреляционной связи между признаками, которая может быть:

* прямой (положительной), Положительная корреляция означает, что при увеличении значений одного признака значения другого признака также увеличиваются.
* обратной (отрицательной), Напротив, отрицательная корреляция свидетельствует о том, что увеличение значений одного признака сопровождается уменьшением значений другого.
* прямолинейной, Прямолинейная связь между переменными проявляется в равномерном изменении средних значений одного признака при одинаковых изменениях другого признака.
* криволинейной. В случае же криволинейной зависимости при равномерном изменении одной переменной можно наблюдать как увеличение, так и уменьшение средних значений другой переменной.
Для измерения и оценки взаимосвязи при прямолинейной корреляции применяется коэффициент корреляции (), а при криволинейной — корреляционное отношение (r).

Значение коэффициента корреляции указывает на силу взаимосвязи между двумя переменными и варьируется от ±1 (полная связь) до 0 (отсутствие связи). Чем ближе значения одного признака к среднему значения другого признака, тем сильнее связь между ними.
Корреляционное отношение (n), или коэффициент детерминации, является важным инструментом для измерения связи между переменными в случае криволинейной зависимости.

В отличие от коэффициента корреляции (r), который применяется для измерения прямолинейной связи, коэффициент детерминации учитывает более сложные формы зависимости и позволяет оценить, насколько одна переменная может объяснить вариацию другой переменной.
*Средняя ошибка корреляции* — это показатель, который используется для оценки точности и достоверности коэффициента корреляции между двумя переменными. Средняя ошибка корреляции представляет собой оценку стандартного отклонения коэффициента корреляции в выборке. Она позволяет оценить, насколько близко измеренная корреляция может соответствовать истинному значению корреляции в генеральной совокупности.
# **Коэффициенты корреляции**
Наличие значимой корреляции между переменными может быть основанием не отвергать гипотезу о возможной причинно-следственной связи. Однако важно помнить, что корреляция сама по себе не является подтверждением причинно-следственной связи. Для подтверждения такой гипотезы требуется дополнительное исследование, включая контроль за другими возможными влияющими факторами и проведение экспериментов, которые позволят установить наличие или отсутствие причинно-следственной связи между переменными.

В медицинских исследованиях применяются различные коэффициенты корреляции, выбор которых определяется типом анализируемых данных. Каждый из этих коэффициентов имеет свои особенности и область применения, что позволяет выбирать наиболее подходящий инструмент для конкретного исследования.

Одним из наиболее часто используемых коэффициентов корреляции в медицинских исследованиях является коэффициент корреляции Пирсона, который позволяет оценить степень линейной взаимосвязи между двумя переменными и лежит в интервале от –1 до 1.
Если коэффициент корреляции равен 1, это означает положительную корреляцию, идеальную прямую линейную зависимость между переменными: увеличение одной переменной сопровождается увеличением другой.

Если коэффициент равен –1, это указывает на отрицательную корреляцию, идеальную обратную линейную зависимость: увеличение одной переменной связано с уменьшением другой.

Коэффициент, близкий к 0, говорит об отсутствии корреляции.

Еще одним важным коэффициентом корреляции является коэффициент корреляции Спирмена. Этот коэффициент также позволяет оценить взаимосвязь между переменными, однако он используется для ранжированных данных и не требует предположений о линейности связи.
Также в медицинских исследованиях могут применяться другие виды коэффициентов корреляции, такие как коэффициент корреляции Кендалла, коэффициент точечной корреляции Фишера и другие, в зависимости от особенностей данных и целей исследования.
# **Коэффициент корреляции Пирсона**
Коэффициент корреляции Пирсона  (линейный коэффициент корреляции) — мера связи для двух непрерывных или характеризующих отношения переменных.
Свойства:
1. Диапазон значений коэффициента r составляет от –1 до +1, что отражает степень силы и направление корреляционной связи между переменными.
2. Знак коэффициента показывает, увеличиваются ли обе переменные вместе (положительная корреляция) или увеличивается одна переменная при уменьшении другой (отрицательная корреляция).
3. Значение коэффициента r отражает, насколько близко расположены точки к прямой линии. При  отсутствует линейная зависимость, хотя может присутствовать нелинейная связь. Чем ближе r к –1 или +1, тем сильнее линейная связь между переменными.
4. Квадрат коэффициента корреляции можно интерпретировать как долю вариации одной переменной, объясненную другой переменной. Например, если , то , что означает, что только 9% изменчивости одной переменной могут быть объяснены изменениями второй переменной.
5. Важное свойство коэффициента корреляции заключается в его устойчивости к изменениям в расположении и масштабе данных. Это означает, что корреляция отражает лишь степень линейной зависимости между двумя переменными, и эта степень не меняется при изменениях в расположении или масштабе данных.
Таким образом, коэффициент корреляции остается надежным инструментом для оценки степени линейной связи между переменными — независимо от того, как данные были трансформированы.
Случаи, когда применение коэффициента линейной корреляции может быть нецелесообразным:
* если между признаками существует нелинейная зависимость, — например, квадратичная;
* если в данных имеется несколько наблюдений по каждому пациенту;
* в случае наличия выбросов или аномальных значений;
* если данные содержат подгруппы пациентов, у которых средние значения по какой-либо из переменных отличаются.
В подобных случаях использование коэффициента линейной корреляции может привести к искаженным результатам или неправильным выводам.

# **Коэффициент корреляции Спирмена**
Коэффициент корреляции Спирмена , также известный как ранговая корреляция, отличается своей универсальностью и неприхотливостью к распределению данных, что делает его особенно ценным для работы с разнообразными типами переменных.

Основная идея коэффициента корреляции Спирмена заключается в том, что данные ранжируются по возрастанию или убыванию, заменяя их на соответствующие ранги. Затем вычисляется степень согласованности между рангами двух переменных, что позволяет определить наличие и силу связи между ними.
