```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def load_data(file_path: str) -> pd.DataFrame:
    """
    Загружает данные из CSV-файла.

    Args:
        file_path (str): Путь к CSV-файлу.

    Returns:
        pd.DataFrame: Загруженные данные.
    """
    return pd.read_csv(file_path, sep=',')

def describe_data(df: pd.DataFrame) -> None:
    """
    Охарактеризует данные, выводя количество наблюдений, количество колонок, имеющиеся колонки,
    основную информацию о данных и статистику по колонкам.

    Args:
        df (pd.DataFrame): DataFrame с данными.
    """
    print("Количество наблюдений:", df.shape[0])
    print("Количество колонок:", df.shape[1])
    print("Имеющиеся колонки:", df.columns)
    print(df.info())
    print(df.describe())

def check_missing_values(df: pd.DataFrame) -> None:
    """
    Проверяет наличие пропущенных значений в DataFrame.

    Args:
        df (pd.DataFrame): DataFrame с данными.
    """
    print(df.isnull().sum())

def visualize_correlation(df: pd.DataFrame) -> None:
    """
    Визуализирует корреляцию между признаками.

    Args:
        df (pd.DataFrame): DataFrame с данными.
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title('Корреляция между признаками')
    plt.show()

def visualize_distribution(df: pd.DataFrame) -> None:
    """
    Визуализирует распределение признаков.

    Args:
        df (pd.DataFrame): DataFrame с данными.
    """
    for column in df.columns[:-1]:  # Исключаем целевую переменную 'Outcome'
        plt.figure(figsize=(10, 6))
        sns.histplot(df[column], kde=True, bins=30)
        plt.title(f'Распределение {column}')
        plt.xlabel(column)
        plt.ylabel('Частота')
        plt.show()

def visualize_distribution_by_target(df: pd.DataFrame) -> None:
    """
    Визуализирует распределение признаков в зависимости от целевой переменной.

    Args:
        df (pd.DataFrame): DataFrame с данными.
    """
    for column in df.columns[:-1]:  # Исключаем целевую переменную 'Outcome'
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='Outcome', y=column, data=df)
        plt.title(f'Распределение {column} в зависимости от Outcome')
        plt.xlabel('Outcome')
        plt.ylabel(column)
        plt.show()

def visualize_target_distribution(df: pd.DataFrame) -> None:
    """
    Визуализирует распределение целевой переменной.

    Args:
        df (pd.DataFrame): DataFrame с данными.
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Outcome', data=df)
    plt.title('Распределение целевой переменной')
    plt.xlabel('Outcome')
    plt.ylabel('Частота')
    plt.show()

def main():
    # Загрузка данных
    di = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\diabetes.csv')
    test = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\test.csv')
    train = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\train.csv')
    di_1 = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\diabetes_012_health_indicators_BRFSS2015.csv')
    da_bi_1 = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
    di_bi_2 = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\diabetes_binary_health_indicators_BRFSS2015.csv')

    # Охарактеризуем данные
    describe_data(di)

    # Проверка на пропущенные значения
    check_missing_values(di)

    # Визуализация корреляции между признаками
    visualize_correlation(di)

    # Визуализация распределения признаков
    visualize_distribution(di)

    # Визуализация распределения признаков в зависимости от целевой переменной
    visualize_distribution_by_target(di)

    # Визуализация распределения целевой переменной
    visualize_target_distribution(di)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2_contingency, ttest_ind

def load_data(file_path: str) -> pd.DataFrame:
    """
    Загружает данные из CSV-файла.

    Args:
        file_path (str): Путь к CSV-файлу.

    Returns:
        pd.DataFrame: Загруженные данные.
    """
    return pd.read_csv(file_path, sep=',')

def check_common_columns(train: pd.DataFrame, test: pd.DataFrame) -> list:
    """
    Проверяет совпадающие колонки между двумя DataFrame.

    Args:
        train (pd.DataFrame): DataFrame с тренировочными данными.
        test (pd.DataFrame): DataFrame с тестовыми данными.

    Returns:
        list: Список совпадающих колонок.
    """
    common_columns = train.columns.intersection(test.columns).tolist()
    print("Совпадающие колонки:", common_columns)
    return common_columns

def merge_data(train: pd.DataFrame, test: pd.DataFrame, common_columns: list) -> pd.DataFrame:
    """
    Объединяет два DataFrame по совпадающим колонкам.

    Args:
        train (pd.DataFrame): DataFrame с тренировочными данными.
        test (pd.DataFrame): DataFrame с тестовыми данными.
        common_columns (list): Список совпадающих колонок.

    Returns:
        pd.DataFrame: Объединенный DataFrame.
    """
    merged_data = pd.merge(train, test, on=common_columns, how='inner')
    print(merged_data.head())
    return merged_data

def clean_data(train: pd.DataFrame, test: pd.DataFrame) -> pd.DataFrame:
    """
    Очищает данные и объединяет их.

    Args:
        train (pd.DataFrame): DataFrame с тренировочными данными.
        test (pd.DataFrame): DataFrame с тестовыми данными.

    Returns:
        pd.DataFrame: Объединенный и очищенный DataFrame.
    """
    # Приведение значений к единому формату
    train['sex'] = train['sex'].str.strip()
    test['sex'] = test['sex'].str.strip()

    # Добавление столбца TenYearCHD в test данные с значением NaN
    test['TenYearCHD'] = np.nan

    # Объединение данных
    all_data = pd.concat([train, test], ignore_index=True)
    print(all_data.head())
    print(all_data.tail())

    # Удаление строк с пропущенными значениями в числовых столбцах
    numeric_features = ['age', 'education', 'cigsPerDay', 'BPMeds', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
    all_data_cleaned = all_data.dropna(subset=numeric_features)

    # Проверка наличия пропущенных значений после очистки
    print(all_data_cleaned.isna().sum())

    return all_data_cleaned

def save_data(df: pd.DataFrame, file_path: str) -> None:
    """
    Сохраняет DataFrame в CSV-файл.

    Args:
        df (pd.DataFrame): DataFrame для сохранения.
        file_path (str): Путь к файлу для сохранения.
    """
    df.to_csv(file_path, index=False)
    print(f"Количество строк в объединенном датасете: {df.shape[0]}")

def visualize_data(df: pd.DataFrame) -> None:
    """
    Визуализирует данные.

    Args:
        df (pd.DataFrame): DataFrame с данными.
    """
    # Визуализация связи между различными факторами и развитием диабета
    plt.figure(figsize=(15, 10))
    sns.pairplot(df, hue='diabetes', vars=['age', 'education', 'BMI', 'glucose', 'TenYearCHD'])
    plt.show()

    # Визуализация распределения диабета по полу
    plt.figure(figsize=(10, 6))
    sns.countplot(x='sex', hue='diabetes', data=df)
    plt.title('Распределение диабета по полу')
    plt.show()

    # Визуализация распределения диабета по возрасту
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='age', hue='diabetes', multiple='stack')
    plt.title('Распределение диабета по возрасту')
    plt.show()

    # Violinplot для распределения признаков (например, давления) среди групп
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='diabetes', y='sysBP', data=df)
    plt.title('Распределение систолического давления по группам с диабетом и без диабета')
    plt.show()

    # Lmplot для линейных зависимостей (например, между возрастом и глюкозой)
    plt.figure(figsize=(10, 6))
    sns.lmplot(x='age', y='glucose', hue='diabetes', data=df)
    plt.title('Линейная зависимость между возрастом и уровнем глюкозы')
    plt.show()

    # Catplot для категориального анализа (например, пол и наличие диабета)
    plt.figure(figsize=(10, 6))
    sns.catplot(x='sex', hue='diabetes', data=df, kind='count')
    plt.title('Категориальный анализ пола и наличия диабета')
    plt.show()

def analyze_data(df: pd.DataFrame, numeric_features: list) -> None:
    """
    Анализирует данные.

    Args:
        df (pd.DataFrame): DataFrame с данными.
        numeric_features (list): Список числовых признаков.
    """
    # Анализ связи между числовыми переменными и диабетом
    for feature in numeric_features:
        diabetes_0 = df[df['diabetes'] == 0][feature]
        diabetes_1 = df[df['diabetes'] == 1][feature]
        t_stat, p_value = ttest_ind(diabetes_0, diabetes_1, equal_var=False)
        print(f'Feature: {feature}, t-statistic: {t_stat}, p-value: {p_value}')

    # Боксплоты для числовых переменных
    for feature in numeric_features:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='diabetes', y=feature, data=df)
        plt.title(f'Распределение {feature} по группам с диабетом и без диабета')
        plt.show()

def main():
    # Загрузка данных
    train = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\train.csv')
    test = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\test.csv')

    # Проверка совпадающих колонок
    common_columns = check_common_columns(train, test)

    # Объединение датасетов по совпадающим колонкам
    merged_data = merge_data(train, test, common_columns)

    # Очистка данных
    all_data_cleaned = clean_data(train, test)

    # Сохранение объединенного датасета в CSV
    save_data(all_data_cleaned, 'all_data.csv')

    # Визуализация данных
    visualize_data(all_data_cleaned)

    # Анализ данных
    numeric_features = ['age', 'education', 'cigsPerDay', 'BPMeds', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
    analyze_data(all_data_cleaned, numeric_features)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def load_data(file_path: str) -> pd.DataFrame:
    """
    Загружает данные из CSV-файла.

    Args:
        file_path (str): Путь к CSV-файлу.

    Returns:
        pd.DataFrame: Загруженные данные.
    """
    return pd.read_csv(file_path, sep=',')

def rename_columns(df: pd.DataFrame, old_name: str, new_name: str) -> pd.DataFrame:
    """
    Переименовывает колонку в DataFrame.

    Args:
        df (pd.DataFrame): DataFrame с данными.
        old_name (str): Старое имя колонки.
        new_name (str): Новое имя колонки.

    Returns:
        pd.DataFrame: DataFrame с переименованной колонкой.
    """
    return df.rename(columns={old_name: new_name})

def combine_dataframes(dfs: list) -> pd.DataFrame:
    """
    Объединяет несколько DataFrame в один.

    Args:
        dfs (list): Список DataFrame для объединения.

    Returns:
        pd.DataFrame: Объединенный DataFrame.
    """
    combined_data = pd.concat(dfs, ignore_index=True)
    combined_data.dropna(inplace=True)
    combined_data.drop_duplicates(inplace=True)
    return combined_data

def visualize_distribution(df: pd.DataFrame, column: str, hue: str, title: str) -> None:
    """
    Визуализирует распределение признака.

    Args:
        df (pd.DataFrame): DataFrame с данными.
        column (str): Имя колонки для визуализации.
        hue (str): Имя колонки для группировки.
        title (str): Заголовок графика.
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column, hue=hue)
    plt.title(title)
    plt.show()

def visualize_boxplot(df: pd.DataFrame, x: str, y: str, title: str) -> None:
    """
    Визуализирует боксплот для признака.

    Args:
        df (pd.DataFrame): DataFrame с данными.
        x (str): Имя колонки для оси X.
        y (str): Имя колонки для оси Y.
        title (str): Заголовок графика.
    """
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=x, y=y)
    plt.title(title)
    plt.show()

def main():
    # Загрузка данных
    di_1 = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\diabetes_012_health_indicators_BRFSS2015.csv')
    da_bi_1 = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
    di_bi_2 = load_data('C:\\IDE\\PYTHON\\итоговый проект\\код\\diabetes_binary_health_indicators_BRFSS2015.csv')

    # Переименование колонок
    di_1 = rename_columns(di_1, 'Diabetes_012', 'Diabetes')
    da_bi_1 = rename_columns(da_bi_1, 'Diabetes_binary', 'Diabetes')
    di_bi_2 = rename_columns(di_bi_2, 'Diabetes_binary', 'Diabetes')

    # Объединение датасетов
    combined_data = combine_dataframes([di_1, da_bi_1, di_bi_2])

    # Проверка результата
    print(combined_data.head())
    print(combined_data.info())

    # Визуализация распределения диабета по возрастным группам
    visualize_distribution(combined_data, 'Age', 'Diabetes', 'Распределение диабета по возрастным группам')

    # Влияние BMI на наличие диабета
    visualize_boxplot(combined_data, 'Diabetes', 'BMI', 'Влияние BMI на наличие диабета')

    # Влияние уровня образования на наличие диабета
    visualize_distribution(combined_data, 'Education', 'Diabetes', 'Влияние уровня образования на наличие диабета')

    # Влияние пола на наличие диабета
    visualize_distribution(combined_data, 'Sex', 'Diabetes', 'Влияние пола на наличие диабета')

    # Влияние физической активности на наличие диабета
    visualize_distribution(combined_data, 'PhysActivity', 'Diabetes', 'Влияние физической активности на наличие диабета')

    # Влияние потребления фруктов и овощей на наличие диабета
    visualize_boxplot(combined_data, 'Diabetes', 'Fruits', 'Влияние потребления фруктов на наличие диабета')
    visualize_boxplot(combined_data, 'Diabetes', 'Veggies', 'Влияние потребления овощей на наличие диабета')

    # Влияние потребления алкоголя на наличие диабета
    visualize_distribution(combined_data, 'HvyAlcoholConsump', 'Diabetes', 'Влияние потребления алкоголя на наличие диабета')

    # Влияние доступа к медицинской помощи на наличие диабета
    visualize_distribution(combined_data, 'AnyHealthcare', 'Diabetes', 'Влияние доступа к медицинской помощи на наличие диабета')

    # Влияние отсутствия доступа к врачу из-за стоимости на наличие диабета
    visualize_distribution(combined_data, 'NoDocbcCost', 'Diabetes', 'Влияние отсутствия доступа к врачу из-за стоимости на наличие диабета')

    # Влияние общего состояния здоровья на наличие диабета
    visualize_boxplot(combined_data, 'Diabetes', 'GenHlth', 'Влияние общего состояния здоровья на наличие диабета')

    # Влияние психического здоровья на наличие диабета
    visualize_boxplot(combined_data, 'Diabetes', 'MentHlth', 'Влияние психического здоровья на наличие диабета')

    # Влияние трудностей при ходьбе на наличие диабета
    visualize_distribution(combined_data, 'DiffWalk', 'Diabetes', 'Влияние трудностей при ходьбе на наличие диабета')

    # Влияние уровня дохода на наличие диабета
    visualize_boxplot(combined_data, 'Diabetes', 'Income', 'Влияние уровня дохода на наличие диабета')

if __name__ == "__main__":
    main()
```
